```TASK 1```

````python
monthdict={
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
````

````python
oldname = 'oct-08_singapore.txt'


parts = oldname.split('_')
mon_day = parts[0]  

month, day = mon_day.split('-')

num_month = monthdict[month.lower()]
newname = f"{num_month}-{day}-{parts[1]}"
print(newname)
````

```TASK 2```

````python
def rename_my_file(old_file_name):
    parts = old_file_name.split('_')
    mon_day = parts[0]  
    month, day = mon_day.split('-')

    num_month = monthdict[month.lower()]
    newname = f"{num_month}-{day}-{parts[1]}"
    return newname
````

````python
#testing function
rename_my_file('oct-08_singapore.txt')
````

````python
folder = 'os-collaboration-exercise-data'
textfiles = glob.glob(os.path.join(folder, "*.txt"))
for oldnamewithpath in textfiles:
    oldname = oldnamewithpath.split(os.path.sep)[-1] #separating the file name from the path, and getting the file name with index -1
    newname = rename_my_file(oldname)
    newpath = os.path.join(folder, newname)        #ensuring all the text files get renamed and stays in the same folder
    os.rename(oldnamewithpath, newpath)
    print(f"{oldname} ----> {newname}")
````

````python
glob.glob('os*/*cairo*')    #calling everything cairo related in the os-collaboration folder
````

````python
allforcairo=glob.glob('os*/*cairo*')      #list of cairo related things 
for file in allforcairo:
    destination = os.path.join('os-collaboration-exercise-data/Cairo')
    shutil.copy(file, destination)        #using shutil to copy (not move)
    print(f'Copied {file} to {destination}')
````

```TASK 7```

````python
for city in citynamelist:                              #calling to previously defined list variable
    print(city)
    allforcity=glob.glob(f'os*/*{city.lower()}*')      #list of resepctive-city related things 
    for file in allforcity:
        destination = os.path.join(f'os-collaboration-exercise-data/{city.lower()}')
        fileindestination = os.path.join(f'os-collaboration-exercise-data/{city.lower()}/{file.split(os.path.sep)[-1]}')
        if os.path.exists(fileindestination):
            print(f'{fileindestination} already exists')
        else:
            shutil.move(file, destination)                 #using shutil to move
            print(f'Moved {file} to {destination}')
````
