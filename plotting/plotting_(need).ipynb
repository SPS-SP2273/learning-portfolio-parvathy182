# 1 A simple plot

## 1.1 Let’s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]       #x-values
y_1 = [0, 2, 4, 6, 8, 10, 12]   #first set of y-values
y_2 = [0, 3, 6, 9, 12, 15, 18]  #another set of y-values

# Let's start plotting
plt.plot(x, y_1, label='Y values',         #plotting x vs y_1, label is the legend, red is the colour of the line and dashed is the line typex
         color='red', linestyle='dashed')
plt.xlabel('x-values')                     #x-axis label
plt.ylabel('y-values')                     #y-axis label
plt.title('X vs Y')                        #graph title
plt.grid(alpha=.25)                        #shows grid lines, the alpha number is the thickness of the gridlines you want
plt.legend(loc='upper right')              #the placement of the legend
```

### Things to note

|Long form|Abbreviation|
|:-|-:|

```python
plt.plot(x, y_2, c='red', ls='dashed', lw=2)
```

## 1.2 Adding another plot

```python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
plt.plot(x, y_2, color='blue', linestyle='dashed', linewidth=2)
plt.xlabel('x-values')                    
plt.ylabel('y-values')                     
plt.title('X vs Y')                        
plt.grid(alpha=.25)                          
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_e, y_e = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_e, yerr=y_e,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 2.1 Adding mathematical functions

Easily generate data on the go, which python lists cannot do

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()          #default is upper left              
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
```

Plot does not look good because sin(x) lies between +/- 1 but x2 has no such boundaries - add another axis!

## We need another axis!

```python
#Using twinx() to add another y-axis
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2             
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
```

### Things to note

```python
x = np.linspace(0, 6, 1000)      #using linspace(start, end, 'how smooth i want it to be? - datapoints')
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # 2 legend calls - 1 For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # 2nd legend call For y-axis 2    


plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
```

# 3 Saving to disc

1. the function savefig() to save to disc
2. specify the format with the extension (e.g., filename.pdf) of the file name
3. specify resolution by use dots-per-inch (dpi)

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)

plt.savefig('simple-plot.png', dpi=150) #nowing saved undr the plotting folder
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)    #small alpha values to soften colour of the plot
plt.ylim([-8, 8])
plt.xlabel('Year')
plt.ylabel('Temperature Anomalies')
plt.title('Temp Anomaly')
plt.grid(alpha=.25)
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)    #small alpha values to soften colour of the plot
plt.ylim([-8, 8])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')         #horizontal line at zero-value
plt.xlabel('Year')
plt.ylabel('Temperature Anomalies')
plt.title('Temp Anomaly')
plt.grid(alpha=.25)
```
